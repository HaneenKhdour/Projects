# Apply 2D-DCT to the matrix

```python

import matplotlib.image as mpimg 
import cv2
 
height, width = 2**6, 2**6
# Read Images  
img = mpimg.imread('pic.jpg') 
img = cv2.resize(img,(height,width))

#show image
plt.figure(figsize=(10,10))
plt.imshow(img) 
plt.show()
```

```python
#Find the zig-zag scan for the matrix
rows=3
columns=3
solution=[[] for i in range(rows+columns-1)]
 
for i in range(rows):
    for j in range(columns):
        sum=i+j
        if(sum%2 ==0):
 
            #add at beginning
            solution[sum].insert(0,data[i][j])
        else:
 
            #add at end of the list
            solution[sum].append(data[i][j])
         
             
# print the solution as it as
for i in solution:
    for j in i:
        print(j,end=" ")
```

```python
#Convert the last K coefficients to zero
rows=3
columns=3
solution=[[] for i in range(rows+columns-1)]
 
for i in range(rows):
    for j in range(columns):
        if i + j > 8:
            data[i,j] = 0.0

plt.imshow(img) 
plt.show()
print(data)
```
